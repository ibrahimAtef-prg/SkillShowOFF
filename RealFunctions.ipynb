# Functions:

**A way of representing inputs and outputs for a certain role or activity—Relationships.** In a function, each input $X$ is manipulated to yield exactly one output $Y$, denoted as:

$$f: X_{in} \rightarrow Y_{out}$$

> ### Mathematical Representation

$$y = f(x)$$

Where:
*   $x$ is the independent variable (input).
*   $y$ is the dependent variable (output).

### Examples of Simple Linear Functions

*   $f(x) = x$
*   $f(x) = x + 5$
*   $f(x) = x - 3$
*   $f(x) = 3x$


SimpleLinearF(x).png

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define the function f(x) = 3x
def f(x):
    return 3 * x

x = np.linspace(-2, 2, 100)
y = f(x)

# Define specific points to highlight : x tweak
x_points = np.array([-1, 1.5, .5])
y_points = f(x_points)

# Create the plot
plt.figure(figsize=(12, 4))
plt.plot(x, y, label='f(x) = 2x + 1', color='blue', zorder=1)

# Scatter specific points
plt.scatter(x_points, y_points, color='red', zorder=2, label='Input_Points(X)')

# Add dotted lines for each input to the axes
for i in range(len(x_points)):
    x_val = x_points[i]
    y_val = y_points[i]

    # Vertical line from x-axis to the point
    plt.plot([x_val, x_val], [0, y_val], color='red', linestyle='-', linewidth=1.2, zorder=0)
    # Horizontal line from y-axis to the point
    plt.plot([0, x_val], [y_val, y_val], color='green', linestyle=':', linewidth=3, zorder=0)

    # Add text annotation
    plt.text(x_val, y_val, f'  x={x_val:.1f}: y={y_val:.1f}', color='black', fontsize=8, ha='right', va='bottom')

plt.title('Simple Function representation')
plt.xlabel('x (Input)')
plt.ylabel('y (Output)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

plt.show()
```

# It's now obvious how *y* changes for each *X* tweak

**But What identify a function? And How?**

: **Vertical Line Test**
VerticalLineTst.png

> - *-GraphBase Test Rule-*
  If a vertical line passes through a graph at more than one point at any time, it is not a function.


: **The Unique Input Rule (Ordered Pairs)**
> - **Rule:** In a set of ordered pairs $(x, y)$, each $x$ (input) must be paired with exactly one $y$ (output).
> *   **Function:** `{(1, 5), (2, 10), (3, 15)}` (All inputs are unique)
> *   **NOT a Function:** `{(1, 5), (1, 10), (2, 15)}` (Input '1' has two different outputs)

### 3. Mapping Diagrams
> **Rule:** In a mapping diagram, every element in the domain (Input) should have exactly **one arrow** pointing to an element in the range (Output).
